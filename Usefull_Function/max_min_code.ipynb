```python
df=pd.read_excel(link)
```

```python
models=list(set(df['Model']))
```

```python
models
```

```python
cnames = list(df.columns) 
print(cnames)
```

```python
def make_func(name_met):
    df_F1_micro=pd.DataFrame()
    for model in models:
    
        df_aux=df[df.Model==model]
        df_aux_bis=df_aux.sort_values(by=[name_met],ascending=False).reset_index(drop=True)

        #df_F1_micro=df_F1_micro.append(df_aux_bis.iloc[[0,-1],:])
        df_F1_micro=df_F1_micro.append(df_aux_bis.iloc[[0],:]) #to have only the max
    return df_F1_micro
```

```python
df_F1_micro = make_func('F1 Micro ')
print(df_F1_micro)
```

```python
df_F1_Macro = make_func('F1 Macro')
df_F1_ACC_MACRO = make_func('ACC Macro')
df_AUNU = make_func('AUNU')
df_AUNP = make_func('AUNP')
```

```python
with pd.ExcelWriter("Max_"+link) as writer:
    df_F1_micro.to_excel(writer,sheet_name='F1_Micro')  
    df_F1_Macro.to_excel(writer, sheet_name='F1 macro')
    df_F1_ACC_MACRO.to_excel(writer, sheet_name='Acc Macro')
    df_AUNU.to_excel(writer, sheet_name='AUNU')
    df_AUNP.to_excel(writer, sheet_name='AUNP')
```
